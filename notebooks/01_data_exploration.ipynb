# 01. Data Exploration & Quality Checks

This notebook performs initial Exploratory Data Analysis (EDA) on the raw UN Comtrade API responses and validates data quality before it enters the ML pipeline.

```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt

# Ensure backend module is importable
sys.path.append(os.path.abspath('../backend'))
from etl.comtrade_transformer import load_raw_json_files, clean_and_transform

plt.style.use('ggplot')
```

## 1. Load & Transform Data

```python
raw_data = load_raw_json_files()
print(f"Raw JSON records loaded: {len(raw_data)}")

if raw_data:
    df = clean_and_transform(raw_data)
    print(f"Transformed records: {len(df)}")
    display(df.head())
else:
    print("No raw data found. Run backend/etl/comtrade_fetcher.py first.")
```

## 2. Data Quality Checks

```python
if raw_data:
    print("Missing Values:")
    print(df.isnull().sum())
    
    print("\nDuplicate Checker (Year, Country, HS Code):")
    dupes = df.duplicated(subset=['year', 'country_iso3', 'hs_code']).sum()
    print(f"Duplicate rows: {dupes}")
```

## 3. High-Level Aggregations

```python
if raw_data:
    # Total value per year
    yearly_val = df.groupby('year')['value_usd'].sum()
    
    plt.figure(figsize=(10, 5))
    yearly_val.plot(kind='line', marker='o')
    plt.title('Total Export Value (USD) by Year')
    plt.ylabel('Value (USD)')
    plt.grid(True)
    plt.show()
```
