## 1. DB Connection and Model Runs Info

```python
db = SessionLocal()
from db.models import ModelRun
runs = db.query(ModelRun).order_by(ModelRun.trained_at.desc()).limit(10).all()
if runs:
    runs_df = pd.DataFrame([{
        'id': r.id,
        'type': r.model_type,
        'trained_at': r.trained_at,
        'mae': r.mae,
        'rmse': r.rmse
    } for r in runs])
    display(runs_df)
else:
    print("No model runs found in DB. Have you trained the models?")
```
