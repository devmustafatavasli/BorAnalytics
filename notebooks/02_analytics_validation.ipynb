# 02. Analytics Validation

This notebook verifies the outputs of the analytics functions: Top Destinations, YoY Growth, and Market Share.

```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt

sys.path.append(os.path.abspath('../backend'))
from db.database import SessionLocal
from analytics.aggregations import get_top_destinations, calculate_yoy_growth
from analytics.market_share import get_market_share

db = SessionLocal()
```

## 1. Top Export Destinations

```python
top_dest = get_top_destinations(db, year=2023, limit=5)
if top_dest:
    df_top = pd.DataFrame(top_dest)
    display(df_top)
    
    df_top.set_index('country_name')['value_usd'].plot(kind='bar')
    plt.title('Top 5 Destinations in 2023')
    plt.ylabel('Value (USD)')
    plt.show()
else:
    print("No data available for 2023. Ensure ETL is run.")
```

## 2. Year-over-Year Growth

```python
yoy = calculate_yoy_growth(db)
if yoy:
    df_yoy = pd.DataFrame(yoy)
    display(df_yoy.tail())
    
    plt.figure(figsize=(10, 4))
    plt.plot(df_yoy['year'], df_yoy['yoy_growth_pct'], marker='o', color='purple')
    plt.axhline(0, color='black', linestyle='--')
    plt.title('YoY Export Growth (%)')
    plt.ylabel('Growth %')
    plt.show()
```

## 3. Market Share Analysis

```python
share = get_market_share(db)
if share:
    df_share = pd.DataFrame(share)
    display(df_share.tail())
    
    df_share.plot(x='year', y=['turkey_share_pct', 'row_share_pct'], kind='bar', stacked=True, figsize=(10, 5))
    plt.title('Turkey vs Rest of World Market Share (%)')
    plt.ylabel('Share %')
    plt.legend(['Turkey', 'ROW'])
    plt.show()
```
